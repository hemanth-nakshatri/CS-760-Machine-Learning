```python
# part 1
import numpy as np

X = np.array([[2, 1, 0],
              [1, 1, 1],
              [0, 1, 0]])

X_inverse = np.linalg.inv(X)
y = np.array([0, 5, 2])
result = np.dot(X_inverse, y)
print("Result:")
print(result)
```

```python
# part 2
w = np.array([0.0, 1.0, 0.0], dtype=float)
learning_rate = 0.02
samples_number = 3


for step in range(500): # w gets closer and closer to [-1, 2, 4]
    # Calculate the gradient (slide 32 from lecture): loss function = (1/samples_number) * ((X*W - y)^2))
    # https://pages.cs.wisc.edu/~jphanna/teaching/2023fall_cs760/slides/lec6-regression-I.pdf

    gradient = (2/samples_number) * (X.T).dot(X.dot(w) - y)
    w -= learning_rate * gradient
    print(f"Step {step + 1}: Weight vector w = {w}")
```
